```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
quake_date = np.genfromtxt('../init_conds/elcentro_NS.dat.txt', delimiter=' ', usecols=(0,1), dtype='f8')
data_damp = np.genfromtxt('../results/build_test.txt', delimiter=',', skip_header=1, usecols=(0,1,2,3,4,5), dtype='f8')
data_no_damp = np.genfromtxt('../results/data_no_damp.txt', delimiter=',', skip_header=1, usecols=(0,1,2,3,4,5), dtype='f8')
```

```python
fig, ax1 = plt.subplots()
fig.set_size_inches(8,5)

ax1.plot(quake_date[:,0], quake_date[:,1])

ax1.set_title('Earth Quake Acceleration, Ground Floor', fontsize=15)
ax1.set_xlabel('Time [s]', fontsize=15)
ax1.set_ylabel('Acceleration [m^2]', fontsize=15)
```

```python
fig, ax1 = plt.subplots()
fig.set_size_inches(8,5)

ax1.plot(quake_date[:,0], data_damp[:,0], label='Floor 1')
ax1.plot(quake_date[:,0], data_damp[:,2], label='Floor 2', alpha=0.75)
# ax1.plot(quake_date, data[:,4], label='Dampener')


ax1.set_title('Dampened Two Story Building', fontsize=15)
ax1.set_xlabel('Time [s]', fontsize=15)
ax1.set_ylabel('Displacement [m]', fontsize=15)

ax1.legend(fontsize=15)
```

```python
fig, ax1 = plt.subplots()
fig.set_size_inches(8,5)

ax1.plot(quake_date[:,0], data_no_damp[:,0], label='Floor 1')
ax1.plot(quake_date[:,0], data_no_damp[:,2], label='Floor 2', alpha=0.75)
# ax1.plot(quake_date, data[:,4], label='Dampener')


ax1.set_title('Dampened Two Story Building', fontsize=15)
ax1.set_xlabel('Time [s]', fontsize=15)
ax1.set_ylabel('Displacement [m]', fontsize=15)

ax1.legend(fontsize=15)
```
