```python
data_tmd = np.genfromtxt('spring_tmd_res.txt', dtype='f8', delimiter=',', usecols=[0,1,2,3])

data_range = np.linspace(0,10, len(data_tmd[:,0]))

fig, ax1 = plt.subplots()
fig.set_size_inches(5,5)

ax1.plot(data_range, data_tmd[:,0], label='Mass 1')
ax1.plot(data_range, data_tmd[:,2], label='Mass 2')

ax1.legend()

ax1.set_title('Position of Spring Masses, Dampning')
ax1.set_xlabel('Time [s]')
ax1.set_ylabel('Mass Position [m]')
```
